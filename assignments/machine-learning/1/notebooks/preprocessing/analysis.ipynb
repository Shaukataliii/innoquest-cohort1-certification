```python
import numpy as np

def calc_lb_ub(column: pd.Series) -> tuple[float | float]:
    q1, q3 = np.quantile(column, [.25, .75])
    iqr = q3 - q1
    lb = q1 - 1.5 * iqr
    ub = q3 + 1.5 * iqr
    return (lb, ub)

lot_area_lb, lot_area_ub = calc_lb_ub(df['Lot Area'])
print()
```
