```python
import pandas as pd
```

```python
df = pd.read_csv("Lending-company.csv")
df.head()
```

### Task 1:

```python
df.drop(['LoanID', 'StringID'], axis=1, inplace=True)
df.head()
```

### Task 2

```python
df['Product'] = df['Product'].map({'Product A': 0, 'Product B': 1, 'Product C': 2, 'Product D': 2, 'Product E': 2, 'Product F': 2})
df.head()
```

### Task 3

```python
df['CustomerGender'].unique()
```

```python
df['CustomerGender'].mode()
```

```python
df['CustomerGender'] = df['CustomerGender'].replace({'NotSpecified': 'Male'})
df['CustomerGender'].unique()
```

### Task 4

```python
import random
idxes = random.sample(range(1000), 5)
df['Location'][idxes]
```

```python
df['Location'] = df['Location'].apply(lambda val: val.split(' ')[-1])
df.head()
```

```python
df['Region'].unique()
```

```python
df['Region'].fillna(df['Region'].mode()[0], inplace=True)
```

```python
df['Region'] = df['Region'].apply(lambda val: val.split(' ')[-1])
df.head()
```

### Task 5

```python
df['TotalPrice'] = df['TotalPrice'].fillna(df['TotalPrice'].mean().round(0))
df['TotalPrice'].isnull().sum()
```

### Task 6

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['StartDate'][df['StartDate'].isna()]
df.drop(index=3, inplace=True)
df.reset_index(drop=True, inplace=True)
df.isnull().sum()
```

```python
df['StartDate'] = pd.to_datetime(df['StartDate'], format="%d/%m/%Y")
```

```python
df['Month'] = [date.month for date in df['StartDate']]
df['Weekday'] = [date.weekday() for date in df['StartDate']]
```

```python
df.head()
```

```python
df.drop('StartDate', axis=1, inplace=True)
df.head()
```

### Task 7

```python
df['TotalDaysYr'].unique()
```

```python
df.drop('TotalDaysYr', axis=1, inplace=True)
df.head()
```

### Task 8

```python
df.dropna(subset=['LoanStatus'], inplace=True)
```

```python
df.isnull().sum()
```
